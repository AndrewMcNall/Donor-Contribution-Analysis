```python
# import libraries needed

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
# read csv files

five_yr_giving = pd.read_csv('NP.5yrGivSum.csv')
giving_2019 = pd.read_csv('NP.givingwoSP.2019.csv')
giving_2018 = pd.read_csv('NP.givingwoSP.2018.csv')
giving_2017 = pd.read_csv('NP.givingwoSP.2017.csv')
giving_2016 = pd.read_csv('NP.givingwoSP.2016.csv')
giving_2015 = pd.read_csv('NP.givingwoSP.2015.csv')
SP_19 = pd.read_csv('NP.SPgiving.2019.csv')
SP_18 = pd.read_csv('NP.SPgiving.2018.csv')
SP_17 = pd.read_csv('NP.SPgiving.2017.csv')
SP_16 = pd.read_csv('NP.SPgiving.2016.csv')
SP_15 = pd.read_csv('NP.SPgiving.2015.csv')
giving_2014 = pd.read_csv('NP_2014_gvg.csv')


# retitle column headings

five_yr_giving.columns = ['ID', 'gvg_w_sc2019', 'gvg_2019', 'gvg_w_sc2018', 'gvg_2018', 'gvg_w_sc2017', 'gvg_2017', 'gvg_w_sc2016', 'gvg_2016',\
                         'gvg_w_sc2015', 'gvg_2015', 'First_Gift', 'FG Date', 'Largest_gift', 'LG_Date', 'Lifetime_w_SC', 'Lifetime_Gift',\
                          'Years_Donated']
giving_2019.columns = ['ID', 'NPo_19', 'NPo_wsc_19', 'LinkedSC', 'SCAmt', 'LinkedName', 'LinkedAmt']
giving_2018.columns = ['ID', 'NPo_18', 'NPo_wsc_18', 'LinkedSC', 'SCAmt', 'LinkedName', 'LinkedAmt']
giving_2017.columns = ['ID', 'NPo_17', 'NPo_wsc_17', 'LinkedSC', 'SCAmt', 'LinkedName', 'LinkedAmt']
giving_2016.columns = ['ID', 'NPo_16', 'NPo_wsc_16', 'LinkedSC', 'SCAmt', 'LinkedName', 'LinkedAmt']
giving_2015.columns = ['ID', 'NPo_15', 'NPo_wsc_15', 'LinkedSC', 'SCAmt', 'LinkedName', 'LinkedAmt']
SP_19.columns = ['ID', 'SP_gvg_2019', 'SP_wsc_2019', 'LinkedSC', 'SCAmt', 'LinkedName', 'LinkedAmt']
SP_18.columns = ['ID', 'SP_gvg_2018', 'SP_wsc_2018', 'LinkedSC', 'SCAmt', 'LinkedName', 'LinkedAmt']
SP_17.columns = ['ID', 'SP_gvg_2017', 'SP_wsc_2017', 'LinkedSC', 'SCAmt', 'LinkedName', 'LinkedAmt']
SP_16.columns = ['ID', 'SP_gvg_2016', 'SP_wsc_2016', 'LinkedSC', 'SCAmt', 'LinkedName', 'LinkedAmt']
SP_15.columns = ['ID', 'SP_gvg_2015', 'SP_wsc_2015', 'LinkedSC', 'SCAmt', 'LinkedName', 'LinkedAmt']
giving_2014.columns = ['ID', 'gvg_2014', 'gvg_w_sc2014']

print(five_yr_giving.head())

print(giving_2014.head())
```

```python
# condense df to columns needed for comparing stats

Combined_gvg = five_yr_giving[['ID', 'gvg_w_sc2019', 'gvg_2019', 'gvg_w_sc2018', 'gvg_2018', 'gvg_w_sc2017', 'gvg_2017', \
                               'gvg_w_sc2016', 'gvg_2016','gvg_w_sc2015', 'gvg_2015', 'Years_Donated']].reset_index(drop = True)

NP_only_2019 = giving_2019[['ID', 'NPo_19', 'NPo_wsc_19']].reset_index(drop = True)
NP_only_2018 = giving_2018[['ID', 'NPo_18', 'NPo_wsc_18']].reset_index(drop = True)
NP_only_2017 = giving_2017[['ID', 'NPo_17', 'NPo_wsc_17']].reset_index(drop = True)
NP_only_2016 = giving_2016[['ID', 'NPo_16', 'NPo_wsc_16']].reset_index(drop = True)
NP_only_2015 = giving_2015[['ID', 'NPo_15', 'NPo_wsc_15']].reset_index(drop = True)
SP_only_2019 = SP_19[['ID', 'SP_gvg_2019', 'SP_wsc_2019']].reset_index(drop = True)
SP_only_2018 = SP_18[['ID', 'SP_gvg_2018', 'SP_wsc_2018']].reset_index(drop = True)
SP_only_2017 = SP_17[['ID', 'SP_gvg_2017', 'SP_wsc_2017']].reset_index(drop = True)
SP_only_2016 = SP_16[['ID', 'SP_gvg_2016', 'SP_wsc_2016']].reset_index(drop = True)
SP_only_2015 = SP_15[['ID', 'SP_gvg_2015', 'SP_wsc_2015']].reset_index(drop = True)
Total_gvg_14 = giving_2014[['ID', 'gvg_2014', 'gvg_w_sc2014']].reset_index(drop = True)
```

```python
# combine df's into a single df to retain empty or 0 values for giving

NP_16to15 = pd.merge(NP_only_2016, NP_only_2015, how = 'outer').reset_index(drop = True)
NP_17to15 = pd.merge(NP_only_2017, NP_16to15, how = 'outer').reset_index(drop = True)
NP_18to15 = pd.merge(NP_only_2018, NP_17to15, how = 'outer').reset_index(drop = True)
NP_19to15 = pd.merge(NP_only_2019, NP_18to15, how = 'outer').reset_index(drop = True)
SP_16to15 = pd.merge(SP_only_2016, SP_only_2015, how = 'outer').reset_index(drop = True)
SP_17to15 = pd.merge(SP_only_2017, SP_16to15, how = 'outer').reset_index(drop = True)
SP_18to15 = pd.merge(SP_only_2018, SP_17to15, how = 'outer').reset_index(drop = True)
SP_19to15 = pd.merge(SP_only_2019, SP_18to15, how = 'outer').reset_index(drop = True)
NP_SP_merge = pd.merge(NP_19to15, SP_19to15, how = 'outer').reset_index(drop = True)
NP_SP_w_2014 = pd.merge(NP_SP_merge, Total_gvg_14, how = 'outer').reset_index(drop = True)
All_merge = pd.merge(Combined_gvg, NP_SP_w_2014, how = 'outer').reset_index(drop = True)

All_merge.fillna(0, inplace = True)
All_merge.to_csv('NP.all_gvg_columns_merge.csv')

print(All_merge.head())
```

```python
# isolate total giving columns as separate lists to remove $0 values.

col_19 = All_merge['gvg_w_sc2019']
nonzero_19 = [amt for amt in col_19 if amt > 0]

col_18 = All_merge['gvg_w_sc2018']
nonzero_18 = [amt for amt in col_18 if amt > 0]

col_17 = All_merge['gvg_w_sc2017']
nonzero_17 = [amt for amt in col_17 if amt > 0]

col_16 = All_merge['gvg_w_sc2016']
nonzero_16 = [amt for amt in col_16 if amt > 0]

col_15 = All_merge['gvg_w_sc2015']
nonzero_15 = [amt for amt in col_15 if amt > 0]

col_14 = All_merge['gvg_w_sc2014']
Count_2014 = np.count_nonzero(col_14)

```

```python
# calculate 2019 Total giving stats

Total_2019 = np.sum(col_19)
Mean_2019 = round(np.mean(nonzero_19), 2)
Median_2019 = np.median(nonzero_19)
Max_2019 = np.max(col_19)
Count_2019 = np.count_nonzero(col_19)
Std_2019 = round(np.std(nonzero_19), 2)

print('2019 stats: Total Gvg - ' + str(Total_2019) + ' Mean: ' + str(Mean_2019) + ' Median: ' + str(Median_2019) + \
     ' Top gift: ' + str(Max_2019) + ' Number of donors: ' + str(Count_2019) + ' Stand Dev: ' + str(Std_2019))

# calculate 2018 Total giving stats

Total_2018 = np.sum(col_18)
Mean_2018 = round(np.mean(nonzero_18), 2)
Median_2018 = np.median(nonzero_18)
Max_2018 = np.max(col_18)
Count_2018 = np.count_nonzero(col_18)
Std_2018 = round(np.std(nonzero_18), 2)

print('2018 stats: Total Gvg - ' + str(Total_2018) + ' Mean: ' + str(Mean_2018) + ' Median: ' + str(Median_2018) + \
     ' Top gift: ' + str(Max_2018) + ' Number of donors: ' + str(Count_2018) + ' Stand Dev: ' + str(Std_2018))

# calculate 2017 Total giving stats

Total_2017 = np.sum(col_17)
Mean_2017 = round(np.mean(nonzero_17), 2)
Median_2017 = np.median(nonzero_17)
Max_2017 = np.max(col_17)
Count_2017 = np.count_nonzero(col_17)
Std_2017 = round(np.std(nonzero_17), 2)

print('2017 stats: Total Gvg - ' + str(Total_2017) + ' Mean: ' + str(Mean_2017) + ' Median: ' + str(Median_2017) + \
     ' Top gift: ' + str(Max_2017) + ' Number of donors: ' + str(Count_2017) + ' Stand Dev: ' + str(Std_2017))

# calculate 2016 Total giving stats

Total_2016 = np.sum(col_16)
Mean_2016 = round(np.mean(nonzero_16), 2)
Median_2016 = np.median(nonzero_16)
Max_2016 = np.max(col_16)
Count_2016 = np.count_nonzero(col_16)
Std_2016 = round(np.std(nonzero_16), 2)

print('2016 stats: Total Gvg - ' + str(Total_2016) + ' Mean: ' + str(Mean_2016) + ' Median: ' + str(Median_2016) + \
     ' Top gift: ' + str(Max_2016) + ' Number of donors: ' + str(Count_2016) + ' Stand Dev: ' + str(Std_2016))

# calculate 2015 Total giving stats

Total_2015 = np.sum(col_15)
Mean_2015 = round(np.mean(nonzero_15), 2)
Median_2015 = np.median(nonzero_15)
Max_2015 = np.max(col_15)
Count_2015 = np.count_nonzero(col_15)
Std_2015 = round(np.std(nonzero_15), 2)

print('2015 stats: Total Gvg - ' + str(Total_2015) + ' Mean: ' + str(Mean_2015) + ' Median: ' + str(Median_2015) + \
     ' Top gift: ' + str(Max_2015) + ' Number of donors: ' + str(Count_2015) + ' Stand Dev: ' + str(Std_2015))
```

```python
# bar graphs on total giving

years = ['2019', '2018', '2017', '2016', '2015']
total_giving = np.array([Total_2019, Total_2018, Total_2017, Total_2016, Total_2015])
mean_array = np.array([Mean_2019, Mean_2018, Mean_2017, Mean_2016, Mean_2015])
median_array = np.array([Median_2019, Median_2018, Median_2017, Median_2016, Median_2015])
total_donors = np.array([Count_2019, Count_2018, Count_2017, Count_2016, Count_2015])


plt.figure(figsize = (12, 8))
plt.subplot(2, 2, 1)
rects1 = plt.bar(years, total_giving, facecolor = 'green', edgecolor = 'black')
plt.title('Total Giving to NP 2015-19 (5 years)')
plt.xlabel('Year')
plt.ylabel('Total $ Amount')

plt.subplot(2, 2, 2)
plt.bar(years, mean_array, facecolor = 'yellow', edgecolor = 'black')
plt.title('Average Giving to NP 2015-19 (5 years)')
plt.xlabel('Year')
plt.ylabel('$ Amount of Average')

plt.subplot(2, 2, 3)
plt.bar(years, median_array, facecolor = 'blue', edgecolor = 'black')
plt.title('Median Giving to NP 2015-19 (5 years)')
plt.xlabel('Year')
plt.ylabel('$ Amount of Median')

plt.subplot(2, 2, 4)
plt.bar(years, total_donors, facecolor = 'purple', edgecolor = 'black')
plt.title('NP Donor Count (5 years)')
plt.xlabel('Year')
plt.ylabel('# of Donors')

plt.subplots_adjust(hspace = 0.35)
plt.savefig('NP_Total_Giving_plots.png')
plt.show()
```

```python
# group donors less $500

sd_2019 = [amt for amt in nonzero_19 if amt < 500]
sd_2018 = [amt for amt in nonzero_18 if amt < 500]
sd_2017 = [amt for amt in nonzero_17 if amt < 500]
sd_2016 = [amt for amt in nonzero_16 if amt < 500]
sd_2015 = [amt for amt in nonzero_15 if amt < 500]

# calculate stats on gifts under $500

or_total_2019 = round(np.sum(sd_2019), 2)
or_total_2018 = round(np.sum(sd_2018), 2)
or_total_2017 = round(np.sum(sd_2017), 2)
or_total_2016 = round(np.sum(sd_2016), 2)
or_total_2015 = round(np.sum(sd_2015), 2)
or_count_2019 = np.count_nonzero(sd_2019)
or_count_2018 = np.count_nonzero(sd_2018)
or_count_2017 = np.count_nonzero(sd_2017)
or_count_2016 = np.count_nonzero(sd_2016)
or_count_2015 = np.count_nonzero(sd_2015)

# group $500-$999 donors

mid_donor_2019 = [amt for amt in nonzero_19 if amt >= 500 and amt < 1000]
mid_donor_2018 = [amt for amt in nonzero_18 if amt >= 500 and amt < 1000]
mid_donor_2017 = [amt for amt in nonzero_17 if amt >= 500 and amt < 1000]
mid_donor_2016 = [amt for amt in nonzero_16 if amt >= 500 and amt < 1000]
mid_donor_2015 = [amt for amt in nonzero_15 if amt >= 500 and amt < 1000]

# calculate stats on $500-$999 gifts

md_total_2019 = round(np.sum(mid_donor_2019), 2)
md_total_2018 = round(np.sum(mid_donor_2018), 2)
md_total_2017 = round(np.sum(mid_donor_2017), 2)
md_total_2016 = round(np.sum(mid_donor_2016), 2)
md_total_2015 = round(np.sum(mid_donor_2015), 2)
md_count_2019 = np.count_nonzero(mid_donor_2019)
md_count_2018 = np.count_nonzero(mid_donor_2018)
md_count_2017 = np.count_nonzero(mid_donor_2017)
md_count_2016 = np.count_nonzero(mid_donor_2016)
md_count_2015 = np.count_nonzero(mid_donor_2015)

# group $1000+ donors

top_donors_2019 = [amt for amt in nonzero_19 if amt >= 1000]
top_donors_2018 = [amt for amt in nonzero_18 if amt >= 1000]
top_donors_2017 = [amt for amt in nonzero_17 if amt >= 1000]
top_donors_2016 = [amt for amt in nonzero_16 if amt >= 1000]
top_donors_2015 = [amt for amt in nonzero_15 if amt >= 1000]

# calculate stats on gifts $1,000+

td_total_2019 = round(np.sum(top_donors_2019), 2)
td_total_2018 = round(np.sum(top_donors_2018), 2)
td_total_2017 = round(np.sum(top_donors_2017), 2)
td_total_2016 = round(np.sum(top_donors_2016), 2)
td_total_2015 = round(np.sum(top_donors_2015), 2)
td_count_2019 = np.count_nonzero(top_donors_2019)
td_count_2018 = np.count_nonzero(top_donors_2018)
td_count_2017 = np.count_nonzero(top_donors_2017)
td_count_2016 = np.count_nonzero(top_donors_2016)
td_count_2015 = np.count_nonzero(top_donors_2015)
```

```python
years = ['2019', '2018', '2017', '2016', '2015']
sm_donors = np.array([or_count_2019, or_count_2018, or_count_2017, or_count_2016, or_count_2015])
md_donors = np.array([md_count_2019, md_count_2018, md_count_2017, md_count_2016, md_count_2015])
td_donors = np.array([td_count_2019, td_count_2018, td_count_2017, td_count_2016, td_count_2015])

print(sm_donors)
print(md_donors)
print(td_donors)

plt.bar(years, sm_donors)
plt.bar(years, md_donors, bottom = sm_donors)
plt.bar(years, td_donors, bottom = sm_donors + md_donors)
plt.legend(['$1-499', '$500-999', '$1000+'])
plt.xlabel('Year')
plt.ylabel('# of Donors')
plt.savefig('stacked_donor_count.png')
plt.show()
```

```python
# calculate stats on retention

# create data frames for each year of donors who gave in year previous and year in variable title

retained_2019 = All_merge[(All_merge.gvg_w_sc2019 > 0)\
                               & (All_merge.gvg_w_sc2018 > 0)].reset_index(drop = True)
retained_2018 = All_merge[(All_merge.gvg_w_sc2018 > 0)\
                               & (All_merge.gvg_w_sc2017 > 0)].reset_index(drop = True)
retained_2017 = All_merge[(All_merge.gvg_w_sc2017 > 0)\
                               & (All_merge.gvg_w_sc2016 > 0)].reset_index(drop = True)
retained_2016 = All_merge[(All_merge.gvg_w_sc2016 > 0)\
                               & (All_merge.gvg_w_sc2015 > 0)].reset_index(drop = True)
retained_2015 = All_merge[(All_merge.gvg_w_sc2015 > 0)\
                                & (All_merge.gvg_w_sc2014 > 0)].reset_index(drop = True)

# calculate number of retained donors

retain_count_2019 = retained_2019.ID.count()
retain_count_2018 = retained_2018.ID.count()
retain_count_2017 = retained_2017.ID.count()
retain_count_2016 = retained_2016.ID.count()
retain_count_2015 = retained_2015.ID.count()


# calculate percentage of retained donors

pct_retention_2019 = round((retain_count_2019 / Count_2018), 3) * 100
pct_retention_2018 = round((retain_count_2018 / Count_2017), 4) * 100
pct_retention_2017 = round((retain_count_2017 / Count_2016), 3) * 100
pct_retention_2016 = round((retain_count_2016 / Count_2015), 3) * 100
pct_retention_2015 = round((retain_count_2015 / Count_2014), 4) * 100

# print results

print('2019 retention: ' + str(pct_retention_2019) + '%')
print(pct_retention_2018)
print(pct_retention_2017)
print(pct_retention_2016)
print(pct_retention_2015)
```

```python
# $500+ retention. md stands for mid-level

# create data frames for each year of donors who gave in year previous and year in variable title

md_retained_2019 = All_merge[(All_merge.gvg_w_sc2019 > 0)\
                               & (All_merge.gvg_w_sc2018 >= 500)].reset_index(drop = True)
md_retained_2018 = All_merge[(All_merge.gvg_w_sc2018 > 0)\
                               & (All_merge.gvg_w_sc2017 >= 500)].reset_index(drop = True)
md_retained_2017 = All_merge[(All_merge.gvg_w_sc2017 > 0)\
                               & (All_merge.gvg_w_sc2016 >= 500)].reset_index(drop = True)
md_retained_2016 = All_merge[(All_merge.gvg_w_sc2016 > 0)\
                               & (All_merge.gvg_w_sc2015 >= 500)].reset_index(drop = True)
md_retained_2015 = All_merge[(All_merge.gvg_w_sc2015 > 0)\
                               & (All_merge.gvg_w_sc2014 >= 500)].reset_index(drop = True)
# print(retained_2019.head())

# counts of donors >= $500

md_count_2018 = (All_merge['gvg_w_sc2018'] >= 500).sum()
md_count_2017 = (All_merge['gvg_w_sc2017'] >= 500).sum()
md_count_2016 = (All_merge['gvg_w_sc2016'] >= 500).sum()
md_count_2015 = (All_merge['gvg_w_sc2015'] >= 500).sum()
md_count_2014 = (All_merge['gvg_w_sc2014'] >= 500).sum()
print(md_count_2018)
print(md_count_2017)

# calculate number of retained donors

md_retain_count_2019 = md_retained_2019.ID.count()
md_retain_count_2018 = md_retained_2018.ID.count()
md_retain_count_2017 = md_retained_2017.ID.count()
md_retain_count_2016 = md_retained_2016.ID.count()
md_retain_count_2015 = md_retained_2015.ID.count()
# print(retain_count_2019)

# calculate percentage of retained donors

pct_retention_md_2019 = round((md_retain_count_2019 / md_count_2018), 3) * 100
pct_retention_md_2018 = round((md_retain_count_2018 / md_count_2017), 4) * 100
pct_retention_md_2017 = round((md_retain_count_2017 / md_count_2016), 2) * 100
pct_retention_md_2016 = round((md_retain_count_2016 / md_count_2015), 3) * 100
pct_retention_md_2015 = round((md_retain_count_2015 / md_count_2014), 4) * 100

# print results

print('Retention $500+ 2019: ' +str(pct_retention_md_2019) + '%')
print('Retention $500+ 2018: ' +str(pct_retention_md_2018) + '%')
print('Retention $500+ 2017: ' +str(pct_retention_md_2017) + '%')
print('Retention $500+ 2016: ' +str(pct_retention_md_2016) + '%')
print('Retention $500+ 2015: ' +str(pct_retention_md_2015) + '%')
```
